# Data Visualization on Covid- 19 Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
sns.set_theme()
```

```python
import mpld3
mpld3.enable_notebook()
```

```python
world = pd.read_csv("data/updated-worldometer")
world.head()
```

```python
world.drop("Unnamed: 0", axis="columns", inplace=True)
```

```python
world.head()
```

```python
world["Population"] = world["Population"].astype(int)
```

### Comparing the TotalCases with TotalRecovered

```python
sns.relplot(x="TotalCases", y="TotalRecovered", data=world, height=8, aspect=10/8);
```

### Comparing the TotalCases with TotalDeaths

```python
sns.relplot(x="TotalCases", y="TotalDeaths", data=world, height=8, aspect=10/8, palette="red");
```

### Comparing the TotalCases with ActiveCases

```python
sns.relplot(x="TotalCases", y="ActiveCases", data=world, height=7,aspect=10/8);
```
